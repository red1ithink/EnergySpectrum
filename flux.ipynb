```python
titles = ["[x1]", "[x10]", "[x0.5]", "[x2]"]
files_list = [files, files2, files3, files4]

for file_group, title in zip(files_list, titles):
    plt.figure()
    for file in file_group:
        label = file.split('/')[-1].split('_')[0]
        ux, uy = sep(file)
        u, v = resizing(ux, uy)
        k_bins, Pi_k = energy_flux(u, v)
        plt.plot(k_bins, Pi_k, label=f"{label}")
        plt.xscale('log')
    plt.title(f'Energy Flux {title}')
    plt.legend()
    plt.show()
```
