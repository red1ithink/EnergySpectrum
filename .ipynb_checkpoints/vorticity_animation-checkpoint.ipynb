```python
import os
import glob
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
```

```python
def numeric_key(fname):
    base = os.path.basename(fname)  
    prefix = base.split('_')[0]   
    return float(prefix)     

folder_path = '../2x/vor'
nx, ny = 1024, 1024     
interval_ms = 500       

file_list = glob.glob(os.path.join(folder_path, "*.txt"))
if len(file_list) == 0:
    raise FileNotFoundError(f"No .txt files found in folder: {folder_path}")

file_list.sort(key=numeric_key)

num_files = len(file_list)
print(f"Found {num_files} text files in {folder_path}.")
data_list = []
for idx, file in enumerate(file_list):
    print(f"Loading file [{idx+1}/{num_files}]: {file}")
    data_1d = np.loadtxt(file, skiprows=1)
    data_2d = data_1d.reshape((nx, ny))
    data_list.append(data_2d)

fig, ax = plt.subplots(figsize=(6, 5))

im = ax.imshow(
    data_list[0],
    origin='lower',
    cmap='jet',
    extent=[0, nx, 0, ny],
    animated=True
)
cbar = plt.colorbar(im, ax=ax)
cbar.set_label("Z-Vorticity")

ax.set_title("Z-Vorticity (Frame=0)")
ax.set_xlabel("X")
ax.set_ylabel("Y")

def init():
    im.set_data(data_list[0])
    ax.set_title("Z-Vorticity (Frame=0)")
    return [im]

def update(frame):
    im.set_data(data_list[frame])
    ax.set_title(f"Z-Vorticity (Frame={frame})")
    return [im]

ani = FuncAnimation(
    fig,
    update,
    frames=num_files,
    init_func=init,
    interval=interval_ms,
    blit=True
)

plt.tight_layout()

save_filename = "vorticity_animation.gif"
ani.save(save_filename, writer="pillow", fps=5)
print(f"Animation saved as {save_filename}")


plt.show()
```
